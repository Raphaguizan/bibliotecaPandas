```python
import pandas as pd
```

```python
aluguelCasa = pd.read_excel('/content/houses_rent.xlsx')
```

```python
aluguelCasa.head(10)
```

```python
aluguelCasa.tail()
```

## Rename

```python
aluguelCasa = aluguelCasa.rename(columns={'parking spaces':'parking_spaces', 'rent amount':'rent_amount', 'property tax':'property_tax', 'fire insurance':'fire_insurance'})
aluguelCasa.head()
```

## Types

```python
aluguelCasa.dtypes
```

```python
aluguelCasa = aluguelCasa.astype({'rent_amount':'float', 'property_tax':'float', 'fire_insurance':'float', 'total':'float'})
aluguelCasa.dtypes
aluguelCasa.head()
```

##Infos

```python
aluguelCasa.info()
aluguelCasa.describe()
```

```python
aluguelCasa[['area','rooms','bathroom','parking_spaces','rent_amount','property_tax','fire_insurance','total']].describe()
```

##Filter

```python
aluguel_casa = aluguelCasa[((aluguelCasa['total'] <= 4000) & (aluguelCasa['animal'] == 'accept')) & ((aluguelCasa['rooms'] == 2) | (aluguelCasa['rooms'] == 3))]
aluguel_casa.describe()
#aluguel_casa.head()
```

```python
aluguel_casa = aluguelCasa.query('total <= 4000 and animal == "accept" and (rooms == 2 or rooms == 3)')
aluguel_casa.describe()
```

##GroupBy

```python
agg = aluguel_casa.groupby(['city']).agg({'total':'mean'})
agg
```

```python
agg_bathroom_price = aluguel_casa.groupby(['bathroom', 'rooms']).agg({'total':'mean'})
agg_bathroom_price
```

##Visualization

### Import libraries (seaborn)

```python
import seaborn as sns
```

###BarChart

```python
ax = sns.barplot(data=agg_bathroom_price, x='rooms', y='total')
ax.set(xlabel='Número de quartos e banheiros', ylabel='Média de preço')
ax.set_title('Total por número de quartos');
```

```python
axB = sns.barplot(data=agg_bathroom_price, x='bathroom', y='total', color="green")
axB.set(xlabel='Número de quartos e banheiros', ylabel='Média de preço')
axB.set_title('Total por número de banheiros');
```

###HistChart

```python
agg_area = aluguel_casa[["area"]]
agg_area = agg_area.query('area <= 100')
sns.histplot(agg_area, kde=True);
```

```python
sns.histplot(aluguel_casa, x='total',hue='city', kde=True);
```
